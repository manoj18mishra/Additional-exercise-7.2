### Exercise 1

    

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
x = np.arange(1,101)
y = x * 2
axes.plot(x, y, 'b')
axes.set_xlabel("x")
axes.set_ylabel("y")
```

```python
z=x**2
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.plot(x, z, 'b')
ax1.set_xlabel("X")
ax1.set_ylabel("Z")

ax2.plot(x, y, 'b')
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_xlim([20.0,22.0])
ax2.set_ylim([30,50])
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(10, 2))
ax[0].plot(x, y, 'b',linewidth=4)
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[1].plot(x, z, 'r--',linewidth=2)
ax[1].set_xlabel("x")
ax[1].set_ylabel("z")
```
